```python
import numpy as np
from scipy.constants import pi, h, k, m_p, m_e, eV
from scipy.optimize import brentq
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
```

```python
def saha(nH,T,fpart):
    """
    nH  := hydrogen density (m**-3)
    T   := temperature [K]
    fpart := 2*U_II/U_I, see expression in LeBlanc
    
    returns n_II*n_e/(n_I*n_H)
    """
    
    # constants
    Eion = 13.6*eV
    kT = k*T
    
    # "quantum" density for electrons
    nQ = (2.0*pi*m_e*kT/h**2)**1.5
    
    return (nQ/nH)*fpart*np.exp(-Eion/kT)

def xion(T,rho,fpart):
    """
    rho  :=  mass density [kg m**-3]
    T   := temperature [K]
    fpart := 2*U_II/U_I, see expression in LeBlanc

    returns ionization fraction x = n_II/(n_I + n_II) of a pure H plasma
    """
    
    # determine the hydrogen density from the mass density
    mH = m_p + m_e
    nH = rho/mH
    s = saha(nH,T,fpart)
    # the ionization fraction is the root of the following quadratic: since 
    # n_e = n_II, n_II**2/n_I = saha(T)
    return 0.5*s*(np.sqrt(1.0+4.0/s) - 1.0)


def boltzmann(T,n,m):
    gn, gm = 2.0*n**2, 2.0*m**2
    dE = (1.0/m**2 - 1.0/n**2)*13.6*eV
    return gn/gm * np.exp(-dE/k/T)

def QH(T):
    terms = np.zeros(200)
    kT_eV = k*T/eV
    for i in range(200):
        E = 13.6*(1.0-1.0/(i+1)**2)
        g = 2.0*(i+1)**2
        terms[i] = g*np.exp(-E/kT_eV)
    return terms.sum()
```

```python
T = np.linspace(3000,30000,270)
q = np.zeros_like(T)
for i in range(len(T)):
    q[i] = QH(T[i])
plt.plot(T,q)
```

```python
rho = 1.0e-6 * 1000.0
fpart = 2.0/q
x = xion(T,rho,fpart)
```

```python
plt.plot(T,1.0-x)
```
